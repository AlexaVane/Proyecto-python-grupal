```python
#!/usr/bin/env python
# -*- coding: utf-8 -*-
from tkinter import Tk,Label,Button,Frame,filedialog
from tkinter import *
import tkinter as tk
import pyaudio
import glob
import os
import wave
import threading
import socket
import numpy as np
from scipy import *
import sys
import math
from gtts import gTTS
from scipy.io.wavfile import read
import matplotlib.pyplot as plt 
from matplotlib.figure import Figure
from matplotlib.backends.backend_tkagg import FigureCanvasTkAgg

grabando=False
reproduciendo=False
CHUNK=1024
data=""
stream=""
f=""
contador=0
contador1=0
contador2=0
port = 5000 #puerto de comunicacion
ip ='172.29.36.37'
def socket_archivo_enviar(ip, port, archivo_audio):
    
    sock = socket.socket(socket.AF_INET, socket.SOCK_STREAM)

    # Realizamos la conexion al la IP y puerto
    sock.connect((ip,port))
    data = sock.recv(4096)
    while True:
        f = open(archivo_audio, "rb")
        content = f.read(1024)
        
        while content:
        # Enviar contenido.
            sock.send(content)
            content = f.read(1024)
        
        break
    #hola
    # Se utiliza el caracter de código 1 para indicar
    # al cliente que ya se ha enviado todo el contenido.
    try:
        sock.send(chr(1))
    except TypeError:
    # Compatibilidad con Python 3.
        sock.send(bytes(chr(1), "utf-8"))
    
    # Cerrar conexión y archivo.
    sock.close()
    f.close()
    

def clear_contador():
    global contador,contador1,contador2
    contador=0
    contador1=0
    contador2=0

def iniciar():
    global grabando
    global proceso
    global act_proceso
    clear_contador()
    audio=pyaudio.PyAudio()
    bloqueo('disabled')
    grabando=True
    FORMAT=pyaudio.paInt16
    CHANNELS=2
    RATE=44100
    act_proceso=True
    archivo="grabacion.wav"
    t1=threading.Thread(target=grabacion, args=(FORMAT,CHANNELS,RATE,CHUNK,audio,archivo))
    t=threading.Thread(target=cuenta)
    t1.start()
    t.start()

def formato(c):
    if c<10:
        c="0"+str(c)
    return c
    
def cuenta():
    global proceso
    global contador,contador1,contador2
    time['text'] = str(formato(contador1))+":"+str(formato(contador2))+":"+str(formato(contador))
    contador+=1
    if contador==60:
        contador=0
        contador2+=1
    if contador2==60:
        contador2=0
        contador1+=1
    proceso=time.after(1000, cuenta)

def abrir():
    global data
    global stream
    global f
    global reproduciendo
    clear_contador()
    audio=pyaudio.PyAudio()
    open_archive=filedialog.askopenfilename(initialdir = "/",
                 title = "Seleccione archivo",filetypes = (("wav files","*.wav"),
                 ("all files","*.*")))
    if open_archive!="":
        reproduciendo=True
        f = wave.open(open_archive,"rb")
        stream = audio.open(format = audio.get_format_from_width(f.getsampwidth()),  
                    channels = f.getnchannels(),  
                    rate = f.getframerate(),
                    output = True)
        data = f.readframes(CHUNK)
        bloqueo('disabled')
        t=threading.Thread(target=cuenta)
        t.start()
        t2=threading.Thread(target=reproduce)
        t2.start()

def reproduce():
    global data
    global stream
    global f
    audio=pyaudio.PyAudio() 
    while data and reproduciendo==True:  
        stream.write(data)  
        data = f.readframes(CHUNK)  
 
    stream.stop_stream()  
    stream.close()  
 
    audio.terminate()
    time.after_cancel(proceso)
    bloqueo('normal')

def bloqueo(s):
    btnIniciar.config(state=s)
    btnDir.config(state=s)
    btnAbrir.config(state=s)

    
def parar():
    global grabando
    global reproduciendo
    if grabando==True:
        grabando=False
        time.after_cancel(proceso)
        clear_contador()
    elif reproduciendo==True:
        reproduciendo=False
    bloqueo('normal')

def direc():
    directorio=filedialog.askdirectory()
    if directorio!="":
        os.chdir(directorio)

def text_to_speech():
    # Funcion para pasar de texto a voz
    from gtts import gTTS # Importamos los modulos necesarios
    shost = texto.get("1.0","end-1c")  # Obtenemos el texto que pasaremos a voz
    print(shost)
    tts = gTTS(text = str(shost), lang='es') # Llamamos al metodo el cual convierto el texto a voz
    archivo_guardar="audio.wav"
    tts.save(archivo_guardar) # Guardamos el audio en un archivo mp3
    socket_archivo_enviar(ip,port,archivo_guardar)

def grabacion(FORMAT,CHANNELS,RATE,CHUNK,audio,archivo):
    
    stream=audio.open(format=FORMAT,channels=CHANNELS,
                          rate=RATE, input=True,
                          frames_per_buffer=CHUNK)

    frames=[]

    while grabando==True:
        data=stream.read(CHUNK)
        frames.append(data)

    #DETENEMOS GRABACIÓN
    stream.stop_stream()
    stream.close()
    audio.terminate()

    #CREAMOS/GUARDAMOS EL ARCHIVO DE AUDIO
    count=0
    for i in glob.glob("*.wav"):
        if "grabacion" in i:
            count+=1
    if count>0:
        archivo="grabacion"+"("+str(count)+")"+".wav"
        
    waveFile = wave.open(archivo, 'wb')
    waveFile.setnchannels(CHANNELS)
    waveFile.setsampwidth(audio.get_sample_size(FORMAT))
    waveFile.setframerate(RATE)
    waveFile.writeframes(b''.join(frames))
    waveFile.close()
    socket_archivo_enviar(ip,port,archivo)

def analisis_audio():
    
    #frame = Frame(window)
        # read audio samples
    open_archive2=filedialog.askopenfilename(initialdir = "/",
                 title = "Seleccione archivo",filetypes = (("wav files","*.wav"),
                 ("all files","*.*")))
    input_data = read(open_archive2)
    window = tk.Toplevel(ventana)
    samplefreq=input_data[0]
    audio = input_data[1]
    # plot the first 1024 samples
    f = Figure(figsize=(5,5), dpi=100)
    a = f.add_subplot(111)
    a.set_title('Señal de audio')
    a.set_xlabel('Tiempo')
    a.set_ylabel('Amplitud')
    a.plot(audio)
    
    duracion_audio = duracion(audio,samplefreq)
    print(duracion_audio)
    
    volumen = volumen_audio(audio,samplefreq)
    print(volumen)
    
    silencios = pausas_rate(audio,samplefreq) 
    print(silencios)
    
    tiempo_hablado = 100 - silencios*100
    
    nivel = nivel_audio(audio,samplefreq)
    print(nivel)
    
    canvas = FigureCanvasTkAgg(f, master=window)
    canvas.get_tk_widget().pack()
    canvas.draw()
    
    w = Canvas(window, 
           width=700,
           height=100)
    w.create_text(350,10,text="Tiempo de diálogo (%)= "+str(tiempo_hablado))
    w.create_text(350,25,text="Duración de grabación (seg)= "+"{0:.2f}".format(duracion_audio))
    w.create_text(350,40,text="Volumen del Locutor (dB)= "+str(volumen))
    w.create_text(350,55,text="Tiempo de pausas del Locutor (%)= " + str(silencios*100))
    w.create_text(350,70,text="Actitud del Locutor= " + nivel)
    w.pack()
    
    
def volumen_audio(archivo_audio,freqSample):
    sonido = archivo_audio
    sampFreq = freqSample
    s1 = sonido[:,0] 
    p = fft(s1)
    n = len(s1)
    nUniquePts = int(math.ceil((n+1)/2.0))
    p = p[0:nUniquePts]
    p = abs(p)
    p = p / float(n) # scale by the number of points so that
                 # the magnitude does not depend on the length 
                 # of the signal or on its sampling frequency  
    p = p**2  # square it to get the power 

    # multiply by two (see technical document for details)
    # odd nfft excludes Nyquist point
    if n % 2 > 0: # we've got odd number of points fft
        p[1:len(p)] = p[1:len(p)] * 2
    else:
        p[1:len(p) -1] = p[1:len(p) - 1] * 2 # we've got even number of points fft

    freqArray = arange(0, nUniquePts, 1.0) * (sampFreq / n)
    decibeles = 10*log10(p)
    volumen_sonido=np.amax(decibeles)
    return volumen_sonido

def pausas_rate(archivo_audio,freqSample):
    snd = archivo_audio[:,0]
    tolerancia = 5
    silencios = sum(1 for i in snd if (i<=(tolerancia) and i>=(-tolerancia)) )#tolerancia a ruido
    rate = silencios/len(snd)
    return rate

def nivel_audio(archivo_audio,freqSample):
    volumen = volumen_audio(archivo_audio,freqSample)
    caracteristica = ''
    if volumen<=(50) and volumen>=(30):
        caracteristica = 'Calmada'
    
    elif volumen>(50):
        caracteristica = 'Agresiva'
        
    elif volumen<(30) and volumen>=(15):
        caracteristica = 'Pasiva'
        
    elif volumen<(15) and volumen>=5:
        caracteristica = 'Adormilado'
        
    else:
        caracteristica = 'Silencio'
        
    return caracteristica
    
def duracion(archivo_audio,freqSample):
    longitud = len(archivo_audio[:,0])
    print(longitud)
    tiempo = longitud/freqSample
    return tiempo
    
ventana = Tk()
ventana.title('Audio')
ventana.geometry("500x500")
ventana.config(bd=15)



lbl_uno = Label(ventana,  relief = GROOVE,anchor=SW,width=450,text="Grabar un audio", font=("Verdana","11")).pack()
frame=Frame(ventana)

time = Label(ventana, relief= RAISED,bd='4',  fg='green',justify='center', width=20, text="00:00:00", bg="black", font=("","30"))
time.pack(pady = 4)
btnIniciar=Button(frame,bd = '3', fg='blue',width=12, text='Iniciar', command=iniciar)
btnIniciar.grid(row=1, column=1,padx=10)
btnParar=Button(frame,bd = '3', fg='blue', width=12, text='Parar', command=parar)
btnParar.grid(row=1, column=2,padx=10)
btnDir=Button(frame,bd = '3', text="Carpeta",width=12,command=direc)
btnDir.grid(row=1,column=3,padx=10)
btnAbrir=Button(frame,bd = '3', text="Abrir",width=12,command=abrir)
btnAbrir.grid(row=1,column=4, padx=4)
frame.pack(pady = 10)

lbl_dos = Label(ventana,relief = GROOVE,anchor=SW,width=450,text="Transformar de texto a voz", font=("Verdana","11"))
lbl_dos.pack(pady = 10)
####################
frame_dos = Frame(ventana)
texto = Text(ventana, relief=RAISED, bd='2')
texto.config(width=450, height=9, font=("Consolas",12), 
             padx=15, pady=15, selectbackground="red")
texto.pack()
btnAceptar=Button(frame_dos,bd = '3', text="Aceptar",width=12,command=lambda:text_to_speech())
btnAceptar.grid(row=1,pady=10, padx=373)
frame_dos.pack()

frame_tres = Frame(ventana)
btnAnalisis=Button(frame_tres,bd = '3', text="Analisis de audio",width=30,command=lambda: analisis_audio())
btnAnalisis.grid(row=1,pady=10, padx=245)
frame_tres.pack()


    

ventana.mainloop()
```
